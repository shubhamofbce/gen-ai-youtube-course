```python
from dotenv import load_dotenv
load_dotenv()
import os
API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from openai import OpenAI
client = OpenAI(api_key=API_KEY)

response = client.responses.create(
  model="gpt-4.1",
  input=[
    {
      "role": "system",
      "content": [
        {
          "type": "input_text",
          "text": "You are a creative and strategic social media assistant that helps brands craft engaging, on-brand social media posts. Your tone should match the brand’s identity (e.g., professional, witty, casual, bold) and adapt based on the platform (e.g., concise and snappy for Twitter, informative and friendly for LinkedIn, visual and energetic for Instagram).\n\nAlways focus on clarity, relevance, and audience engagement. Suggest hashtags, emojis (if appropriate), and strong calls-to-action. Make sure posts align with marketing goals like promoting products, increasing brand awareness, announcing events, or driving traffic.\n\nAsk clarifying questions if the brand’s tone, product, or campaign goal isn’t clear."
        }
      ]
    },
    {
      "role": "user",
      "content": [
        {
          "type": "input_text",
          "text": "My brand is Nike, Can you write a good instagram post for me"
        }
      ]
    }
  ],
  text={
    "format": {
      "type": "text"
    }
  },
  reasoning={},
  tools=[],
  temperature=1.51,
  max_output_tokens=2048,
  top_p=1,
  store=True
)

print(response.output_text)
```
